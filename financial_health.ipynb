```python
import pandas as pd

df = pd.read_csv("/content/financial_data.csv")

total_revenue = df["revenue"].sum()
total_expense = df["expense"].sum()
profit = total_revenue - total_expense

print("Total Revenue:", total_revenue)
print("Total Expense:", total_expense)
print("Profit:", profit)
```

```python
if profit > 0:
    print("‚úÖ The business is PROFITABLE.")
else:
    print("‚ö†Ô∏è The business is in LOSS.")

expense_ratio = (total_expense / total_revenue) * 100
print("Expense Ratio:", round(expense_ratio, 2), "%")

if expense_ratio > 70:
    print("‚ö†Ô∏è Expenses are too high. Cost optimization needed.")
else:
    print("‚úÖ Expenses are under control.")
```

```python
def generate_ai_insights(total_revenue, total_expense, profit):
    summary = ""
    risks = []
    recommendations = []

    expense_ratio = (total_expense / total_revenue) * 100

    if profit < 0:
        summary = "The business is running at a loss."
        risks.append("Continuous losses may affect business sustainability.")
        recommendations.append("Reduce unnecessary expenses.")
    else:
        summary = "The business is financially healthy and profitable."
        recommendations.append("Reinvest profits to expand the business.")

    if expense_ratio > 70:
        risks.append("High expense ratio detected.")
        recommendations.append("Optimize operational costs.")
    else:
        risks.append("Expenses may increase if not monitored closely.")
        recommendations.append("Continue monitoring expenses regularly.")

    risks.append("Lack of emergency cash reserve.")
    recommendations.append("Maintain 3‚Äì6 months of emergency funds.")

    return summary, risks, recommendations

```

```python
summary, risks, recommendations = generate_ai_insights(
    total_revenue, total_expense, profit
)

print("üß† AI Financial Health Summary:")
print(summary)

print("\n‚ö†Ô∏è Risks:")
for r in risks:
    print("-", r)

print("\n‚úÖ Recommendations:")
for rec in recommendations:
    print("-", rec)
```
